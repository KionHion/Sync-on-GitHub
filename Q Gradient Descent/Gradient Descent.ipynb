## 自己寫Gradient Descent
函數: $x^4+5\left(x+1\right)^2$
<br> 微分後:     $4x^3+10(x+1)$

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

這個是“設計”好的函數

```python
x=np.linspace(-1.5,0.1,100)
y=(x**4)+(5*(x+1)**2)
plt.plot(x,y);
```

```python
def f(x):
    return (x**4)+(5*(x+1)**2)
```

```python
def dydx(x):
    return 4*x**3+10*(x+1)
```

畫出來比較直觀

```python
x=np.linspace(-1.5,0.5,100)
y=(x**4)+(5*(x+1)**2)
plt.plot(x,y);

InitialPoint=-5
x=InitialPoint
μ=0.01
y_new=[]
Ex=[]
for i in range(10000000000000000000):
  #  print(x)
    x1=x-(μ*dydx(x))
    Ex.append(x1)
    y_new.append(f(x1))
    if abs(x1-x)<0.00000001:
        minX=x1
      #  return(minX,i)
        break
    else:
        x=x1

for i in range(20):
    plt.scatter(Ex[i],y_new[i])
```

## Interact
看$\mu$的大小會如何影響需要算多少次

```python
def minimum_x(μ):
    InitialPoint=0.8
    x=InitialPoint
    for i in range(10000000000000000000):
        x1=x-(μ*dydx(x))
        if abs(x1-x)<0.00000000001:
            minX=x1
            return(i)
            break
        else:
            x=x1
```

```python
from ipywidgets import interact

interact(minimum_x,μ=(0.01,0.11,0.001))
```
