## 介紹：

* 瘧疾是一種寄生蟲傳染病
* *Plasmodium falciparum* 會透過瘧蚊寄生在人類身上，而其中一環是寄生在人類的紅血球上
* 檢驗方式就是透過染劑觀察紅血球有沒有*Plasmodium falciparum*
 *[Source](https://en.azvision.az/news/63804/world-malaria-day-2017-what-is-malaria,-symptoms-and-what-today-is-all-about.html) *



## 實做方法：
資料來源透過[Kaggle](https://www.kaggle.com/iarunava/cell-images-for-detecting-malaria?fbclid=IwAR1m-jzGwGi9FtR7KtnD7LG_FNDaEXWA-XVLIt2XwIGdaqoqEpdNF7s5J-c)取得的細胞圖片，分別是患有瘧疾跟沒有患病的細胞，然後利用CNN和NN進行圖像分析，分析細胞是否患有瘧疾，透過各種測試，來比較一下，找出比較好的模型。
